# NYC Taxi Traffic Analysis



Number of NYC taxi passengers, where the five anomalies occur during the NYC marathon, Thanksgiving, Christmas, New Years day, and a snow storm. The raw data is from the NYC Taxi and Limousine Commission. The data file included here consists of aggregating the total number of taxi passengers into 30 minute buckets.

In this notebook, I will provide a study on the taxi traffic in NYC using [Anomaly Detection Toolkit](https://adtk.readthedocs.io/en/stable/) (ADTK), a Python package for **unsupervised / rule-based time series anomaly detection**.

As the nature of anomaly varies over different cases, a model may not work universally for all anomaly detection problems. Choosing and combining detection algorithms (detectors), feature engineering methods (transformers), and ensemble methods (aggregators) properly is the key to build an effective anomaly detection model.

This package offers a set of common detectors, transformers and aggregators with unified APIs, as well as pipe classes that connect them together into a model. It also provides some functions to process and visualize time series and anomaly events.

Several questions will be answered during this study:
1. What is the frequence of the pattern for the traffic in NYC?
2. What anomalies can be detected in the historic data?
3. What informations can we tell about this anomaly?
4. How can we use those data to detect anomalies in new data?

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import adtk
```

```python
from adtk.data import validate_series

df = pd.read_csv("data/dataset.csv",usecols=["timestamp","value"], index_col="timestamp", parse_dates=True,squeeze=True)
df = pd.DataFrame(df)
df.index = pd.to_datetime(df.index)
ts = validate_series(df)
ts.rename(columns = {"value":"Traffic"},inplace = True)
ts.index.rename("Time",inplace = True)
```

```python
ts
```

```python
ts_train = ts.loc[ts.index<pd.to_datetime("2015-01-01 00:00:00")]
ts_test = ts.loc[ts.index>=pd.to_datetime("2015-01-01 00:00:00")]
```

We have 7 months of data.

```python
from adtk.visualization import plot
plot(ts_train)
```

## What is the frequence of the pattern for the traffic in NYC?

```python
frequencies={
    "hourly":2, # because two units in one hour
    "dayly":2*24,
    "weekly":2*24*7,
    "monthly":2*24*30}
```

```python
from adtk.transformer import ClassicSeasonalDecomposition

s_transformed_by_frequency = ts_train

for key,frequence in frequencies.items():
    s_transformed = ClassicSeasonalDecomposition(freq=frequence).fit_transform(ts_train).rename(columns={"Traffic": "Seasonal decomposition residual for a "+key+"-frequency"})
    sum_residuals = abs(s_transformed).sum()
    
    print("A seasonal decomposition residual for a "+key+"-frequency gives a total of "+str(sum_residuals[0])+" of residuals.")
    
    s_transformed_by_frequency = pd.concat([s_transformed_by_frequency, s_transformed], axis=1)
```

```python
plot(s_transformed_by_frequency, ts_markersize=1)
```

> 📌 Seeing the sums of residuals and the plots, we can tell that a **weekly frequency** is the best way to deal with this time-series. That makes sense since the general use of taxis will probably depend on the the day.

----

## What anomalies can be detected in the historical data?

```python
from adtk.detector import SeasonalAD
seasonal_ad = SeasonalAD(c=5.0, side="both")
anomalies = seasonal_ad.fit_detect(ts_train)
plot(ts_train, anomaly=anomalies, anomaly_color="red")
```

```python
from adtk.data import to_events
anomalies_events = to_events(anomalies)["Traffic"]
nb_anomalies = len(anomalies_events)
print(str(nb_anomalies)+" anomalies have been detected.")
```

----

## What informations can we tell about those anomalies?

In this part I will try to describe the anomalies by defining different metrics to characterize them.

### 1. The deviation
The deviation corresponds to the percentage of variation the anomaly refers to, compared to usual values we are supposed to have.

First we can determine the pattern calculated, and then we can use this pattern to compute the rate of deviation between the pattern and the residuals.

```python
residuals = ClassicSeasonalDecomposition().fit_transform(ts_train)
pattern = ts_train - residuals
plot(pattern, ts_markersize=1);
```

```python
def deviation(ts, residuals, event):
    pattern = ts - residuals
    subset_pattern = pattern.loc[(pattern.index>=event[0])&(pattern.index<=event[1])]
    sum_pattern = subset_pattern.sum()
    
    subset_residuals = residuals.loc[(residuals.index>=event[0])&(residuals.index<=event[1])]
    sum_residuals = subset_residuals.sum()
    
    return int(round(sum_residuals/sum_pattern*100))
```

```python
deviation_ = deviation(ts_train, residuals, anomalies_events[0])
print("For the first anomaly, we have a deviation of "+str(deviation_)+"%.")
```

### 2. The normal traffic
The normal traffic corresponds to the mean of taxis we were supposed to have during the anomaly.

```python
def normal_traffic(ts, residuals, event):
    pattern = ts - residuals
    subset_pattern = pattern.loc[(pattern.index>=event[0])&(pattern.index<=event[1])]
    
    return int(round(subset_pattern.mean()))
```

```python
normal_traffic_ = normal_traffic(ts_train, residuals, anomalies_events[0])
print("For the first anomaly, we have a normal traffic of "+str(normal_traffic_)+" taxis.")
```

### 3. The duration of the anomaly

```python
def duration(event):
    delta = event[1] - event[0]
    
    return str(delta)
```

```python
duration_ = duration(anomalies_events[0])
print("For the first anomaly, we have a duration of "+duration_+".")
```

### 4. The frequency of anomalies
Here we will define at what frequency we have anomalies.

```python
def frequency(ts, anomalies):
    delta = ts.index.max() - ts.index.min()
    return round(len(anomalies)/delta.days*30,2)
```

```python
frequency_ = frequency(ts_train, anomalies_events)
print("There are  "+str(frequency_)+" anomalies a month.")
```

### Characterization of the anomalies

```python
recap_anomalies = pd.DataFrame(columns=['Start', 'End', 'Duration', 'Deviation', 'Normal Traffic', 'Way'])

for event in anomalies_events:
    deviation_ = deviation(ts_train, residuals, event)
    normal_traffic_ = normal_traffic(ts_train, residuals, event)
    duration_ = duration(event)
    
    way = 'UP' if deviation_>0 else 'DOWN'
    
    data = {'Start': [str(event[0])],
        'End': [str(event[1])],
        'Duration': [duration_],
        'Deviation': [deviation_],
        'Normal Traffic': [normal_traffic_],
        'Way': [way]}
    
    new_row = pd.DataFrame(data)
    recap_anomalies = pd.concat([recap_anomalies, new_row],ignore_index=True)
    
frequency_ = frequency(ts_train, anomalies_events)
print("There are in average "+str(frequency_)+" anomalies a month.")
```

```python
recap_anomalies
```

> 📌 We can see with this recap that the some anomalies seem to impact more the Traffic than other with higher deviation or impacting a more important number of taxis.

## How can we use those data to detect anomalies in new data?

```python
anomalies_test = seasonal_ad.predict(ts_test)
anomalies_events_test = to_events(anomalies_test)["Traffic"]
plot(ts_test, anomaly=anomalies_test, anomaly_color="red")
```

> 📌 Using the historical data we can then provide an alerting system for anomalies in the traffic in real time.
